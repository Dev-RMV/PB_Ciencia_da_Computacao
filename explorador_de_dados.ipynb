```python
# Preparação dos dados para comparação

# Criar pasta para salvar as figuras
pasta_figuras = "./graficos"
if not os.path.exists(pasta_figuras):
    os.makedirs(pasta_figuras)

# Configurações para figuras de alta qualidade para paper
plt.rcParams.update({
    'figure.dpi': 300,
    'savefig.dpi': 300,
    'savefig.format': 'png',
    'savefig.bbox': 'tight',
    'savefig.pad_inches': 0.2,
    'font.size': 10,
    'axes.titlesize': 12,
    'axes.labelsize': 11,
    'xtick.labelsize': 10,
    'ytick.labelsize': 10,
    'legend.fontsize': 10,
    'figure.titlesize': 14
})

# Criar um dicionário com todos os DataFrames e suas respectivas estruturas
estruturas_dados = {
    'Lista Encadeada': {'linux': df_LE_linux, 'windows': df_LE_windows},
    'Lista Duplamente Encadeada': {'linux': df_LDE_linux, 'windows': df_LDE_windows},
    'Lista Encadeada Circular': {'linux': df_LEC_linux, 'windows': df_LEC_windows},
    'Lista Duplamente Encadeada Circular': {'linux': df_LDEC_linux, 'windows': df_LDEC_windows}
}

# Consolidar todos os dados em um único DataFrame
dados_consolidados = []

for estrutura, plataformas in estruturas_dados.items():
    for plataforma, df in plataformas.items():
        df_temp = df.copy()
        df_temp['estrutura_dados'] = estrutura
        df_temp['plataforma'] = plataforma
        dados_consolidados.append(df_temp)

# Combinar todos os dados
df_completo = pd.concat(dados_consolidados, ignore_index=True)

print(f"Dados consolidados: {len(df_completo)} registros")
print(f"Estruturas de dados: {df_completo['estrutura_dados'].unique()}")
print(f"Plataformas: {df_completo['plataforma'].unique()}")
print(f"Operações: {df_completo['operacao'].unique()}")
```

```python
# Calcular tempo total de execução por operação, estrutura e plataforma
tempo_total_por_grupo = df_completo.groupby(['plataforma', 'estrutura_dados', 'operacao'])['tempo_execucao_ms'].sum().reset_index()
tempo_total_por_grupo['tempo_execucao_s'] = tempo_total_por_grupo['tempo_execucao_ms'] / 1000

print("Análise do tempo total por grupo:")
print(tempo_total_por_grupo.head(10))
```

```python
# Configuração do estilo dos gráficos
plt.style.use('default')
sns.set_palette("husl")

# Figura 1: Comparação geral por operação (soma de todas as estruturas)
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Comparação de Performance: Windows vs Linux\n(Tempo Total de Execução por Operação)', fontsize=16, fontweight='bold')

# Calcular tempo total por operação e plataforma (somando todas as estruturas)
tempo_por_operacao = df_completo.groupby(['plataforma', 'operacao'])['tempo_execucao_ms'].sum().reset_index()

# Gráfico de barras agrupadas
ax1 = axes[0, 0]
sns.barplot(data=tempo_por_operacao, x='operacao', y='tempo_execucao_ms', hue='plataforma', ax=ax1)
ax1.set_title('Tempo Total por Operação (todas as estruturas)', fontweight='bold')
ax1.set_xlabel('Operação')
ax1.set_ylabel('Tempo Total (ms)')
ax1.tick_params(axis='x', rotation=45)
ax1.legend(title='Plataforma')

# Gráfico de barras com diferença percentual
tempo_pivot = tempo_por_operacao.pivot(index='operacao', columns='plataforma', values='tempo_execucao_ms').fillna(0)
tempo_pivot['diferenca_percentual'] = ((tempo_pivot['windows'] - tempo_pivot['linux']) / tempo_pivot['linux'] * 100)

ax2 = axes[0, 1]
cores = ['red' if x > 0 else 'green' for x in tempo_pivot['diferenca_percentual']]
tempo_pivot['diferenca_percentual'].plot(kind='bar', ax=ax2, color=cores)
ax2.set_title('Diferença Percentual: Windows vs Linux\n(Positivo = Windows mais lento)', fontweight='bold')
ax2.set_xlabel('Operação')
ax2.set_ylabel('Diferença (%)')
ax2.tick_params(axis='x', rotation=45)
ax2.axhline(y=0, color='black', linestyle='-', alpha=0.3)

# Gráfico de dispersão - Comparação direta
ax3 = axes[1, 0]
scatter_data = tempo_pivot.reset_index()
ax3.scatter(scatter_data['linux'], scatter_data['windows'], s=100, alpha=0.7)
ax3.plot([0, max(scatter_data[['linux', 'windows']].max())],
         [0, max(scatter_data[['linux', 'windows']].max())],
         'k--', alpha=0.5, label='Linha de igualdade')
ax3.set_xlabel('Tempo Linux (ms)')
ax3.set_ylabel('Tempo Windows (ms)')
ax3.set_title('Correlação Direta: Linux vs Windows', fontweight='bold')
ax3.legend()

# Adicionar labels nos pontos
for i, txt in enumerate(scatter_data['operacao']):
    ax3.annotate(txt, (scatter_data['linux'].iloc[i], scatter_data['windows'].iloc[i]),
                xytext=(5, 5), textcoords='offset points', fontsize=8)

# Box plot comparativo
ax4 = axes[1, 1]
sns.boxplot(data=df_completo, x='plataforma', y='tempo_execucao_ms', ax=ax4)
ax4.set_title('Distribuição dos Tempos de Execução', fontweight='bold')
ax4.set_xlabel('Plataforma')
ax4.set_ylabel('Tempo de Execução (ms)')

plt.tight_layout()

# Salvar figura 1
fig.savefig(f'{pasta_figuras}/figura_01_comparacao_geral.png', dpi=300, bbox_inches='tight',
            facecolor='white', edgecolor='none')
fig.savefig(f'{pasta_figuras}/figura_01_comparacao_geral.pdf', dpi=300, bbox_inches='tight',
            facecolor='white', edgecolor='none')
plt.show()
print(f"✓ Figura 1 salva em {pasta_figuras}/")
```

```python
# Figura 2: Comparação detalhada por estrutura de dados
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Comparação Detalhada por Estrutura de Dados\n(Windows vs Linux)', fontsize=16, fontweight='bold')

estruturas = df_completo['estrutura_dados'].unique()

for i, estrutura in enumerate(estruturas):
    ax = axes[i//2, i%2]

    # Filtrar dados para a estrutura atual
    dados_estrutura = tempo_total_por_grupo[tempo_total_por_grupo['estrutura_dados'] == estrutura]

    # Criar gráfico de barras agrupadas
    sns.barplot(data=dados_estrutura, x='operacao', y='tempo_execucao_ms', hue='plataforma', ax=ax)
    ax.set_title(f'{estrutura}', fontweight='bold')
    ax.set_xlabel('Operação')
    ax.set_ylabel('Tempo Total (ms)')
    ax.tick_params(axis='x', rotation=45)
    ax.legend(title='Plataforma')

plt.tight_layout()

# Salvar figura 2
fig.savefig(f'{pasta_figuras}/figura_02_comparacao_detalhada.png', dpi=300, bbox_inches='tight',
            facecolor='white', edgecolor='none')
fig.savefig(f'{pasta_figuras}/figura_02_comparacao_detalhada.pdf', dpi=300, bbox_inches='tight',
            facecolor='white', edgecolor='none')
plt.show()
print(f"✓ Figura 2 salva em {pasta_figuras}/")
```

```python
# Figura 3: Heatmap de comparação
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 8))
fig.suptitle('Heatmap de Performance: Tempo Total por Operação e Estrutura', fontsize=16, fontweight='bold')

# Criar tabelas pivot para cada plataforma
pivot_linux = tempo_total_por_grupo[tempo_total_por_grupo['plataforma'] == 'linux'].pivot(
    index='estrutura_dados', columns='operacao', values='tempo_execucao_ms').fillna(0)

pivot_windows = tempo_total_por_grupo[tempo_total_por_grupo['plataforma'] == 'windows'].pivot(
    index='estrutura_dados', columns='operacao', values='tempo_execucao_ms').fillna(0)

# Heatmap Linux
sns.heatmap(pivot_linux, annot=True, fmt='.2f', cmap='YlOrRd', ax=ax1, cbar_kws={'shrink': 0.8})
ax1.set_title('Linux - Tempo Total (ms)', fontweight='bold')
ax1.set_xlabel('Operação')
ax1.set_ylabel('Estrutura de Dados')

# Heatmap Windows
sns.heatmap(pivot_windows, annot=True, fmt='.2f', cmap='YlOrRd', ax=ax2, cbar_kws={'shrink': 0.8})
ax2.set_title('Windows - Tempo Total (ms)', fontweight='bold')
ax2.set_xlabel('Operação')
ax2.set_ylabel('Estrutura de Dados')

plt.tight_layout()

# Salvar figura 3
fig.savefig(f'{pasta_figuras}/figura_03_heatmap_comparacao.png', dpi=300, bbox_inches='tight',
            facecolor='white', edgecolor='none')
fig.savefig(f'{pasta_figuras}/figura_03_heatmap_comparacao.pdf', dpi=300, bbox_inches='tight',
            facecolor='white', edgecolor='none')
plt.show()
print(f"✓ Figura 3 salva em {pasta_figuras}/")
```

```python
# Figura 4: Análise de eficiência relativa
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Análise de Eficiência Relativa por Estrutura de Dados', fontsize=16, fontweight='bold')

for i, estrutura in enumerate(estruturas):
    ax = axes[i//2, i%2]

    # Filtrar e pivotar dados para a estrutura atual
    dados_estrutura = tempo_total_por_grupo[tempo_total_por_grupo['estrutura_dados'] == estrutura]
    pivot_estrutura = dados_estrutura.pivot(index='operacao', columns='plataforma', values='tempo_execucao_ms').fillna(0)

    # Calcular razão Windows/Linux
    pivot_estrutura['razao_win_linux'] = pivot_estrutura['windows'] / pivot_estrutura['linux']

    # Gráfico de barras horizontais
    cores = ['red' if x > 1 else 'green' for x in pivot_estrutura['razao_win_linux']]
    pivot_estrutura['razao_win_linux'].plot(kind='barh', ax=ax, color=cores)
    ax.set_title(f'{estrutura}\n(Razão Windows/Linux)', fontweight='bold')
    ax.set_xlabel('Razão (>1 = Windows mais lento)')
    ax.set_ylabel('Operação')
    ax.axvline(x=1, color='black', linestyle='--', alpha=0.5, label='Igualdade')
    ax.legend()

plt.tight_layout()

# Salvar figura 4
fig.savefig(f'{pasta_figuras}/figura_04_eficiencia_relativa.png', dpi=300, bbox_inches='tight',
            facecolor='white', edgecolor='none')
fig.savefig(f'{pasta_figuras}/figura_04_eficiencia_relativa.pdf', dpi=300, bbox_inches='tight',
            facecolor='white', edgecolor='none')
plt.show()
print(f"✓ Figura 4 salva em {pasta_figuras}/")
```

```python
# Figura 5: Gráfico especializado para paper - Comparação de performance normalizada
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))
fig.suptitle('Performance Comparison: Linux vs Windows Data Structures', fontsize=14, fontweight='bold')

# Preparar dados para o gráfico especializado
tempo_medio_por_grupo = df_completo.groupby(['plataforma', 'estrutura_dados', 'operacao'])['tempo_execucao_ms'].mean().reset_index()

# Gráfico 1: Tempo médio por estrutura (agrupado por plataforma)
tempo_por_estrutura = df_completo.groupby(['plataforma', 'estrutura_dados'])['tempo_execucao_ms'].mean().reset_index()

sns.barplot(data=tempo_por_estrutura, x='estrutura_dados', y='tempo_execucao_ms', hue='plataforma', ax=ax1)
ax1.set_title('Average Execution Time by Data Structure', fontweight='bold')
ax1.set_xlabel('Data Structure')
ax1.set_ylabel('Average Time (ms)')
ax1.tick_params(axis='x', rotation=45)
ax1.legend(title='Platform')

# Gráfico 2: Performance normalizada (Linux como baseline = 1.0)
pivot_normalizado = tempo_por_estrutura.pivot(index='estrutura_dados', columns='plataforma', values='tempo_execucao_ms')
pivot_normalizado['windows_normalized'] = pivot_normalizado['windows'] / pivot_normalizado['linux']

cores_norm = ['red' if x > 1 else 'green' for x in pivot_normalizado['windows_normalized']]
pivot_normalizado['windows_normalized'].plot(kind='bar', ax=ax2, color=cores_norm, alpha=0.7)
ax2.set_title('Windows Performance Relative to Linux\n(Linux baseline = 1.0)', fontweight='bold')
ax2.set_xlabel('Data Structure')
ax2.set_ylabel('Normalized Performance Ratio')
ax2.tick_params(axis='x', rotation=45)
ax2.axhline(y=1, color='black', linestyle='--', alpha=0.5, label='Linux baseline')
ax2.legend()

# Adicionar valores nas barras
for i, v in enumerate(pivot_normalizado['windows_normalized']):
    ax2.text(i, v + 0.02, f'{v:.2f}', ha='center', va='bottom', fontweight='bold')

plt.tight_layout()

# Salvar figura 5 (especializada para paper)
fig.savefig(f'{pasta_figuras}/figura_05_paper_performance_comparison.png', dpi=300, bbox_inches='tight',
            facecolor='white', edgecolor='none')
fig.savefig(f'{pasta_figuras}/figura_05_paper_performance_comparison.pdf', dpi=300, bbox_inches='tight',
            facecolor='white', edgecolor='none')
plt.show()
print(f"✓ Figura 5 (especializada para paper) salva em {pasta_figuras}/")
```

```python
# Figura 6: Matriz de correlação para paper
fig, ax = plt.subplots(1, 1, figsize=(10, 8))

# Criar matriz de dados para correlação
matriz_correlacao = tempo_total_por_grupo.pivot_table(
    index=['estrutura_dados', 'operacao'],
    columns='plataforma',
    values='tempo_execucao_ms',
    fill_value=0
)

# Calcular correlação entre Linux e Windows
correlacao = matriz_correlacao['linux'].corr(matriz_correlacao['windows'])

# Scatter plot com linha de regressão
scatter = ax.scatter(matriz_correlacao['linux'], matriz_correlacao['windows'],
                    alpha=0.7, s=60, c='steelblue', edgecolors='black', linewidth=0.5)

# Adicionar linha de regressão
z = np.polyfit(matriz_correlacao['linux'], matriz_correlacao['windows'], 1)
p = np.poly1d(z)
ax.plot(matriz_correlacao['linux'], p(matriz_correlacao['linux']), "r--", alpha=0.8, linewidth=2)

# Linha de igualdade
max_val = max(matriz_correlacao[['linux', 'windows']].max())
ax.plot([0, max_val], [0, max_val], 'k--', alpha=0.3, linewidth=1, label='Perfect correlation')

ax.set_xlabel('Linux Execution Time (ms)', fontweight='bold')
ax.set_ylabel('Windows Execution Time (ms)', fontweight='bold')
ax.set_title(f'Platform Performance Correlation\n(r = {correlacao:.3f})', fontweight='bold')
ax.legend()
ax.grid(True, alpha=0.3)

# Adicionar texto com estatísticas
stats_text = f'Correlation: r = {correlacao:.3f}\nSamples: n = {len(matriz_correlacao)}'
ax.text(0.05, 0.95, stats_text, transform=ax.transAxes, fontsize=10,
        verticalalignment='top', bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8))

plt.tight_layout()

# Salvar figura 6
fig.savefig(f'{pasta_figuras}/figura_06_correlation_matrix.png', dpi=300, bbox_inches='tight',
            facecolor='white', edgecolor='none')
fig.savefig(f'{pasta_figuras}/figura_06_correlation_matrix.pdf', dpi=300, bbox_inches='tight',
            facecolor='white', edgecolor='none')
plt.show()
print(f"✓ Figura 6 (matriz de correlação) salva em {pasta_figuras}/")
```

```python
# Tabela resumo com estatísticas (salvar como CSV para uso no paper)
print("="*80)
print("RESUMO ESTATÍSTICO DA COMPARAÇÃO WINDOWS vs LINUX")
print("="*80)

# Estatísticas gerais
stats_gerais = df_completo.groupby('plataforma')['tempo_execucao_ms'].agg([
    'count', 'mean', 'median', 'std', 'min', 'max'
]).round(4)

print("\n1. ESTATÍSTICAS GERAIS:")
print(stats_gerais)

# Salvar estatísticas gerais
stats_gerais.to_csv(f'{pasta_figuras}/tabela_01_estatisticas_gerais.csv')

# Comparação por operação
print("\n2. TEMPO TOTAL POR OPERAÇÃO (ms):")
comparacao_operacao = tempo_por_operacao.pivot(index='operacao', columns='plataforma', values='tempo_execucao_ms')
comparacao_operacao['diferenca_abs'] = comparacao_operacao['windows'] - comparacao_operacao['linux']
comparacao_operacao['diferenca_perc'] = (comparacao_operacao['diferenca_abs'] / comparacao_operacao['linux'] * 100).round(2)
print(comparacao_operacao)

# Salvar comparação por operação
comparacao_operacao.to_csv(f'{pasta_figuras}/tabela_02_comparacao_operacoes.csv')

# Matriz de dados completa para análises adicionais
tempo_total_por_grupo.to_csv(f'{pasta_figuras}/dados_completos_analise.csv', index=False)

print(f"\n✓ Todas as tabelas salvas em {pasta_figuras}/")
print(f"✓ Dados completos salvos para análises adicionais")

# Resumo das estruturas mais eficientes
resumo_eficiencia = []
print("\n3. ESTRUTURA MAIS EFICIENTE POR OPERAÇÃO:")
for operacao in df_completo['operacao'].unique():
    dados_op = tempo_total_por_grupo[tempo_total_por_grupo['operacao'] == operacao]
    melhor_linux = dados_op[dados_op['plataforma'] == 'linux'].nsmallest(1, 'tempo_execucao_ms')
    melhor_windows = dados_op[dados_op['plataforma'] == 'windows'].nsmallest(1, 'tempo_execucao_ms')

    if not melhor_linux.empty and not melhor_windows.empty:
        print(f"{operacao}:")
        print(f"  Linux: {melhor_linux['estrutura_dados'].iloc[0]} ({melhor_linux['tempo_execucao_ms'].iloc[0]:.2f} ms)")
        print(f"  Windows: {melhor_windows['estrutura_dados'].iloc[0]} ({melhor_windows['tempo_execucao_ms'].iloc[0]:.2f} ms)")

        resumo_eficiencia.append({
            'operacao': operacao,
            'melhor_estrutura_linux': melhor_linux['estrutura_dados'].iloc[0],
            'tempo_linux_ms': melhor_linux['tempo_execucao_ms'].iloc[0],
            'melhor_estrutura_windows': melhor_windows['estrutura_dados'].iloc[0],
            'tempo_windows_ms': melhor_windows['tempo_execucao_ms'].iloc[0]
        })

# Salvar resumo de eficiência
pd.DataFrame(resumo_eficiencia).to_csv(f'{pasta_figuras}/tabela_03_resumo_eficiencia.csv', index=False)

print(f"\n{'='*80}")
print(f"🎯 RESUMO DA EXPORTAÇÃO:")
print(f"📁 Pasta criada: {pasta_figuras}/")
print(f"🖼️  6 figuras salvas em PNG (300 DPI) e PDF")
print(f"📊 4 tabelas CSV salvas para análises")
print(f"📈 Dados completos exportados")
print(f"✅ Pronto para uso em paper acadêmico!")
print(f"{'='*80}")
```
